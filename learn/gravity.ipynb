# Gravitational Variations

Load ephemeris data calculated by [JPL Horizons](http://ssd.jpl.nasa.gov/?horizons) for the following configurations:
* Sun center relative to solar system barycenter
* Moon center relative to solar system barycenter
* Earth center relative to solar system barycenter
* Clock relative to Earth center

The empheris for each configuration is calculated as (x,y,z) in km and (vx,vy,vz) in km/s.

```python
# Define a function to read the csv-like emphemeris output from Horizon and return a numpy recarray
# with fields jdct,x,y,z,vx,vy,vz. Units are days, km, and km/s.
def loadEphemeris(name,path = '/Users/david/Documents/Personal/Clocks/Gears/learn/gravity/'):
    records = [ ]
    skipHeader = True
    for line in open(path + name,'r'):
        line = line.rstrip()
        if line == '$$SOE':
            skipHeader = False
            continue
        if line == '$$EOE':
            break
        if skipHeader:
            continue
        tokens = line.split(',')
        row = [ ]
        for i in (0,2,3,4,5,6,7):
            row.append(float(tokens[i]))
        records.append(row)
    print 'Read %d lines from %s' % (len(records),name);
    raw = np.array(records);
    return raw.view(dtype=[('jdct',float),('x',float),('y',float),('z',float),('vx',float),('vy',float),('vz',float)]).flat.copy()
```

```python
sun = loadEphemeris('sun.txt');
moon = loadEphemeris('moon.txt');
earth = loadEphemeris('earth.txt');
irvine = loadEphemeris('irvine.txt');
```

```python
# Create a time variable that starts at zero and measures days by offsetting the emphemeris JDCT field.
t_days = sun['jdct'] - np.min(sun['jdct']);
```

```python
# Define a function that returns an array of 3D position vectors from the results of loadEphemeris()
def pos(ephem):
    return ephem[['x','y','z']].view(dtype=float).reshape((-1,3))
```

```python
# Define a function that returns an array of 3D velocity vectors from the results of loadEphemeris()
def vel(ephem):
    return ephem[['vx','vy','vz']].view(dtype=float).reshape((-1,3))
```

```python
# Define a function that calculates the array of magnitudes to the specified power for an array of 3-vectors.
def mag(vecs,pow=1.):
    return np.sum(vecs*vecs,axis=1)**(pow/2.)
```

```python
# Define a function that returns a vector with magnitude 1/|r1-r2|^2 and the same direction as r2-r1.
# If r2 is None, then r2 = 0 is assumed. Inputs of r1,r2 are assumed to be km.
def invSquare(r1,r2=None):
    if r2 is not None:
        dr = r2 - r1;
    else:
        dr = -r1;
    return dr/mag(dr,3)[:,np.newaxis]/(1e3*units.m)**2
```

```python
# Physical constants are values with units, so define the units we expect for calculated accelerations.
ms2 = units.meter/units.second**2;
```

```python
# Calculate position and velocity vectors in the solar system reference frame for
# the sun, moon, and earth centers of gravity. Units are km and km/s (but dimensions are not attached).
r_sun = pos(sun);
r_moon = pos(moon);
r_earth = pos(earth);
#r_pivot =  r_earth - pos(irvine);
v_sun = vel(sun);
v_moon = vel(moon);
v_earth = vel(earth);
#v_pivot =  v_earth - vel(irvine);
```

```python
# Calculate the unit vector pointing from the pivot to the earth's center of gravity, which
# defines "down" for measuring theta.
r_pivot = pos(irvine);
down = r_pivot/mag(r_pivot)[:,np.newaxis];
```

```python
# Calculate the unit vector pointing eastwards from the pivot.
v_pivot = -vel(irvine);
east = v_pivot/mag(v_pivot)[:,np.newaxis];
```

```python
# Verify that the down and east unit vectors are orthogonal, as required since earth moves as a rigid body.
assert np.allclose(np.sum(down*east,axis=-1),0.),'Down and East unit vectors are not orthogonal!'
```

```python
# Calculate the north unit vector that completes the down,east,north right-handed coordinate system.
north = np.cross(down,east);
```

```python
# Check that unit vectors are normalized.
rtol,atol= 1e-10,1e-14
assert np.allclose(mag(down),1.,rtol=rtol,atol=atol),'Down unit vectors are not normalized!'
assert np.allclose(mag(east),1.,rtol=rtol,atol=atol),'East unit vectors are not normalized!'
assert np.allclose(mag(north),1.,rtol=rtol,atol=atol),'North unit vectors are not normalized!'
```

```python
class VectorSpline(object):
    """
    An object representing a vector function of time and its time derivative
    """
    def __init__(self,vec,t=t_days,k=3):
        self.components = [ ]
        self.derivatives = [ ]
        for axis in range(vec.shape[-1]):
            spline = InterpolatedUnivariateSpline(t,vec[:,axis],k=k)
            self.components.append(spline)
            self.derivatives.append(spline.derivative(n=1))
    def __call__(self,tvec,deriv=False):
        """
        Evaluates the vector function (or its derivative if deriv is True) at each time in the input tvec array.
        """
        result = np.empty((len(tvec),3))
        for axis,interpolator in enumerate(self.derivatives if deriv else self.components):
            result[:,axis] = interpolator(tvec)
        return result
```

```python
def var(x):
    x_mag = mag(x)
    x_avg = np.mean(x_mag)
    return 1e6*(x_mag - x_avg)
```
