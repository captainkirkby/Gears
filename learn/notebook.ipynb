```python
# Load required modules
import math
from sklearn import linear_model,gaussian_process
from scipy import sparse
```

```python
# Load 5 days of samples in medium-rare format
dump = np.load('/Volumes/Data/clock/sampleData.npy');
nSamples = len(dump);
print 'Loaded %d samples' % (nSamples);
# Convert sample number to days.
tdays = np.arange(nSamples)/(24*3600.);
```

```python
# Define a function to calculate variations of a signal about the signal mean and
# optionally print statistics
def getVariation(signal,label='',unit='',scale=1,printStats=False):
    mean = np.mean(signal)
    variation = scale*(signal - mean)
    if printStats:
        print '%s mean = %f %s, rms = %f %s, range = [%+f,%+f] %s' % (
            label,scale*mean,unit,np.std(variation),unit,np.min(variation),np.max(variation),unit)
    return variation
```

```python
# Calculate variations and print statistics for the input data
periodVariation = getVariation(dump['refinedPeriod'],'Period','us',scale=1e6,printStats=True);
amplitudeVariation = getVariation(dump['angle'],'Amplitude','deg',printStats=True);
blockTempVariation = getVariation(dump['blockTemperature'],'Block temp','C',printStats=True);
boardTempVariation = getVariation(dump['boardTemperature'],'Board temp','C',printStats=True);
pressureVariation = getVariation(dump['pressure'],'Air pressure','hPa',scale=1e-2,printStats=True);
relhumidVariation = getVariation(dump['humidity'],'Relative humidity','%',printStats=True);
```

```python
# Superimpose two hours of period and amplitude data separated by 24 hours, which shows that:
# - The dominant variations (due to gear train errors) are periodic on this timescale.
# - The period varies most over a few minutes while the amplitude variations are slower.
# - The period and amplitude are affected differently by environmental conditions.
nhour = 3600;
t1,t2 = 2*nhour,26*nhour;
thour = 60.*np.arange(nhour)/nhour;
plt.figure(figsize=(12,10));
plt.subplot(2,1,1);
plt.plot(thour,periodVariation[t1:t1+nhour],'-',label='hour-1');
plt.plot(thour,periodVariation[t2:t2+nhour],'-',label='hour-2');
plt.xlabel('Elapsed time (mins)');
plt.ylabel('Period variation (us)');
plt.legend();
plt.subplot(2,1,2);
plt.plot(thour,amplitudeVariation[t1:t1+nhour],'-',label='hour-1');
plt.plot(thour,amplitudeVariation[t2:t2+nhour],'-',label='hour-2');
plt.xlabel('Elapsed time (mins)');
plt.ylabel('Amplitude variation (deg)');
plt.legend();
```

```python
# Define a function to perform a running average of a signal. Use odd length values to
# avoid half-sample shift. The edges of the signal (where there is not enough data on
# both sides for a running average of the requested length) are not smoothed at all.
def runningAvg(signal,length):
    kernel = np.ones((length,))/length
    smooth = np.copy(signal)
    i1 = (length-1)//2
    i2 = len(signal) - length + i1 + 1
    smooth[i1:i2] = np.convolve(signal,kernel,mode='valid')
    return smooth
```

```python
# Smooth the environmental parameters with a 55-sec running average.
nsmooth = 55;
blockTempSmooth = runningAvg(dump['blockTemperature'],nsmooth);
boardTempSmooth = runningAvg(dump['boardTemperature'],nsmooth);
pressureSmooth = runningAvg(dump['pressure'],nsmooth);
relhumidSmooth = runningAvg(dump['humidity'],nsmooth);
```

```python
# Plot an hour of environmental variations with and without smoothing, which shows that:
# - Board is about 7 degC warmer than block, presumably due to local heating.
# - Block temperature is more stable than board temperature, as expected due to
#   block's increased thermal mass.
# - Honeywell sensors have some readout aliasing (horizontal banding), but this is
#   effectively removed by the running average.
t1 = 6*nhour;
plt.figure(figsize=(12,12));
plt.subplot(2,2,1);
plt.plot(thour,dump['blockTemperature'][t1:t1+nhour],'r.');
plt.plot(thour,blockTempSmooth[t1:t1+nhour],'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Block Temperature (C)');
plt.subplot(2,2,2);
plt.plot(thour,dump['boardTemperature'][t1:t1+nhour],'r.');
plt.plot(thour,boardTempSmooth[t1:t1+nhour],'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Board Temperature (C)');
plt.subplot(2,2,3);
plt.plot(thour,dump['pressure'][t1:t1+nhour]-1e5,'r.');
plt.plot(thour,pressureSmooth[t1:t1+nhour]-1e5,'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Air Pressure - 100,000 (Pa)');
plt.subplot(2,2,4);
plt.plot(thour,dump['humidity'][t1:t1+nhour],'r.');
plt.plot(thour,relhumidSmooth[t1:t1+nhour],'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Relative Humidity (%)');
```

```python
# Use the smoothed block temperature as our best guess of the ambient air temperature.
airTemperature = blockTempSmooth;
```

```python
# Convert the smoothed relative humidity signal into a partial pressure of water vapor.
# The first step is to calculate the saturated water vapor pressure (in Pa) as a function
# of the ambient air pressure and temperature (see http://en.wikipedia.org/wiki/Relative_humidity)
saturatedVaporPressure = 1e2*(1.0007 + 3.46e-8*pressureSmooth)*6.1121*np.exp(17.502*airTemperature/(240.97+airTemperature));
# Next, we scale the saturated pressure by the relative humidity percentage.
vaporPressure = 1e-2*relhumidSmooth*saturatedVaporPressure;
# Plot the results.
plt.figure(figsize=(12,5));
plt.plot(tdays,saturatedVaporPressure,label='saturated');
plt.plot(tdays,vaporPressure,label='actual');
plt.xlabel('Elapsed Time (days)');
plt.ylabel('Water vapor pressure (Pa)');
plt.ylim((0,4000));
plt.legend();
```

```python
# Calculate the dry air partial pressure as the difference between the total and vapor pressures,
# assuming that air can be treated as an ideal gas.
dryPressure = pressureSmooth - vaporPressure;
# Values of the specific gas constants Rx = kB/mx for dry air and water vapor in J/(kg K)
Rdry,Rvapor = 287.058,461.495;
# Calculate the air density in g/m^3 using the ideal gas law.
airDensity = 1e3*(dryPressure/Rdry + vaporPressure/Rvapor)/(airTemperature+273.15);
# Plot the results.
plt.figure(figsize=(12,5));
plt.plot(tdays,airDensity);
plt.xlabel('Elapsed Time (days)');
plt.ylabel('Air Pressure (g/m^3)');
```

```python
# Model the train errors as an arbitrary additive signal with a period of exactly 12*60*60 samples.
# Since the signal is arbitrary, each of the 12*60*60 possible phases is described by an independent
# parameter, so our train error feature matrix T selects the appropriate parameter for each sample.
# This is a large matrix whose elements are mostly zeros with a small fraction of ones, so we use
# a sparse format to keep the memory requirements low.
trainErrorPeriod = 12*60*60;
ntrainCycles = (nSamples+trainErrorPeriod-1)//trainErrorPeriod;
# Build a sparse identity matrix that represents one cycle of train errors.
ident = sparse.identity(trainErrorPeriod,dtype=np.float64);
# Repeat for the total number of cycles in the data, rounded up to a complete cycle.
T = sparse.vstack([ident]*ntrainCycles);
# Trim rows beyond the actual number of samples that were added by rounding up.
T = T.tocsr()[:nSamples];
```

```python
# Use variations in the air density, water vapor pressure, and smoothed block temperature to model
# environmental influcences on the period and amplitude.
airDensityVariation = getVariation(airDensity);
vaporPressureVariation = getVariation(vaporPressure);
blockTempSmoothVariation = getVariation(blockTempSmooth);
X = np.vstack([airDensityVariation,vaporPressureVariation,blockTempSmoothVariation]).T;
```

```python
# Combine the train error model T and the environmental model X, using a sparse representation
# for the result XT.
TX = sparse.hstack([T,sparse.csr_matrix(X)]);
# Make a version of the same model with environmental variations turned off.
T0 = sparse.hstack([T,sparse.csr_matrix(np.zeros_like(X))]);
```

$$
\delta y(t,x_1,x_2,\ldots) = \delta y_T(t\ \text{mod}\ T) + \sum_i \delta x_i \frac{\partial y}{\partial x_i}
$$

```python
# Define a function for learning model parameters for an arbitrary signal
# and returning a 'cooked' decomposition of the signal into its model components.
# If plots is True, then generate a set of reference plots.
def learn(signal,label=None,envLabels=('Air Density (g/m^3)','Vapor Pressure (Pa)','Temperature (C)'),
          plots=True,noiseSeqLen=240):
    nEnv = len(envLabels);
    assert nEnv+trainErrorPeriod == TX.shape[1], 'Unexpected number of envLabels';
    # Initialize the model.
    model = linear_model.LinearRegression();
    # Fit the model parameters to the signal.
    model.fit(TX,signal);
    # Calculate the best-fit model and residuals.
    fit = model.predict(TX);
    residuals = signal - fit;
    # Print environmental parameter linear coefficients.
    envCoefs = model.coef_[trainErrorPeriod:];
    print 'Environmental linear coeficients:',envCoefs;
    # Calculate the predicted variations due to train errors only.
    trainErrors = model.predict(T0);
    print 'Train error range:',(np.min(trainErrors),np.max(trainErrors));
    # Subtract the train errors to obtain the purely environmental variations.
    envFit = fit - trainErrors;
    # Calculate the seperate contributions from each environmental parameter.
    envEach = (envCoefs*X).T;
    # Estimate the random noise component of the residuals by calculating the RMS
    # of independent sequences.
    nSeq = len(signal)//noiseSeqLen;
    residualsRMS = np.std(residuals[:nSeq*noiseSeqLen].reshape((-1,noiseSeqLen)),axis=1);
    # Use the 10th percentile of these RMS values as our noise estimate.
    noiseRMS = np.percentile(residualsRMS,10);
    # Perform a running average smoothing of the residuals with the same sequence length.
    residualsSmooth = runningAvg(residuals,noiseSeqLen);
    # Estimate the random noise by subtracting off the smooth residuals.
    noise = residuals - residualsSmooth;
    # Calculate the truncated RMS of this noise as a cross check.
    limits = np.percentile(noise,(2,98));
    truncated = np.logical_and(noise > limits[0],noise < limits[1]);
    noiseTruncRMS = np.std(noise[truncated]);
    print 'Noise RMS = %f (%f)' % (noiseRMS,noiseTruncRMS);
    if plots:
        # Initialize plots.
        plt.figure(figsize=(12,12));
        # Plot one hour of data to highlight train errors.
        plt.subplot(4,1,1);
        plt.plot(thour,trainErrors[t1:t1+nhour],'b',label='Train Errors');
        plt.plot(thour,envFit[t1:t1+nhour],'r',label='Environmental');
        plt.plot(thour,residuals[t1:t1+nhour],alpha=0.3,label='Residuals');
        plt.xlabel('Elapsed Time (mins)');
        if label:
            plt.ylabel(label);
        plt.legend();
        # Plot full dataset to highlight environmental contributions.
        plt.subplot(4,1,2);
        plt.plot(tdays,envEach[0],label='Air Density');
        plt.plot(tdays,envEach[1],label='Humidity');
        plt.plot(tdays,envEach[2],label='Temperature');
        ylim = plt.gca().get_ylim();
        plt.ylim(ylim);
        plt.plot(tdays,residualsSmooth,label='Residuals');
        plt.plot(tdays,noise,alpha=0.25,label='Noise');
        plt.xlabel('Elapsed Time (days)');
        if label:
            plt.ylabel(label);
        plt.legend();
        # Plot variation wrt to each environmental parameter to test linearity.
        for i in range(nEnv):
            plt.subplot(4,nEnv,2*nEnv+i+1);
            x = X[:,-(nEnv-i)];
            xlim = np.array(np.percentile(x,(1,99)));
            y = residuals+envEach[i];
            ylim = np.percentile(y,(1,99));
            plt.plot(x,y,'.',markersize=1,alpha=0.2);
            plt.plot(xlim,envCoefs[i]*xlim,'r-',label='%.3g' % envCoefs[i]);
            plt.xlim(xlim);
            plt.ylim(ylim);
            plt.xlabel(envLabels[i]);
            plt.ylabel(label);
            plt.legend();
        # Show residuals noise analysis plots.
        rmsRange = (0.,np.percentile(residualsRMS,99));
        plt.subplot(4,3,10);
        plt.plot(tdays[:nSeq*noiseSeqLen:noiseSeqLen],residualsRMS,'+');
        plt.ylim(rmsRange);
        plt.xlabel('Elapsed Time (days)');
        plt.ylabel('Residuals RMS');
        plt.subplot(4,3,11);
        plt.hist(residualsRMS,bins=100,range=rmsRange);
        plt.xlabel('Residuals RMS');
        plt.subplot(4,3,12);
        plt.hist(noise[truncated],bins=100);
        plt.xlabel('Residuals - Smooth');
    # Return the environmental coefficients and the cooked dataset.
    return envCoefs,np.vstack((signal,trainErrors,residualsSmooth)).T;
```

```python
periodCoefs,periodCooked = learn(periodVariation,label='Period Variation (us)');
```

```python
amplitudeCoefs,amplitudeCooked = learn(amplitudeVariation,label='Amplitude Variation (deg)');
```

```python
# Combined the period and amplitude cooked results into a single dataset to save.
# Note that we do not include any explicit time stamp column to avoid the need to
# generate ISO date strings which are redundant with the sample index.
cooked = np.vstack((
    np.arange(nSamples).astype(float),
    airDensityVariation,vaporPressureVariation,blockTempSmoothVariation,
    periodCooked.T,amplitudeCooked.T)).T.copy();
# Add column labels.
cooked = cooked.view(dtype=[('index',float),('airDensity',float),('vaporPressure',float),('temperature',float),
                ('period',float),('periodTrainError',float),('periodResidual',float),
                ('amplitude',float),('amplitudeTrainError',float),('amplitudeResidual',float)
                ]).flat.copy();
```
