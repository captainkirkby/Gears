```python
# Load 5 days of samples in medium-rare format
dump = np.load('/Volumes/Data/clock/sampleData.npy');
nSamples = len(dump);
print 'Loaded %d samples' % (nSamples);
# Convert sample number to days.
tdays = np.arange(nSamples)/(24*3600.);
```

```python
# Calculate variations of a signal about the signal mean and optionally print statistics
def getVariation(signal,label='',unit='',scale=1,printStats=False):
    mean = np.mean(signal)
    variation = scale*(signal - mean)
    if printStats:
        print '%s mean = %f %s, rms = %f %s, range = [%+f,%+f] %s' % (
            label,scale*mean,unit,np.std(variation),unit,np.min(variation),np.max(variation),unit)
    return variation
```

```python
# Calculate variations of the input signals
periodVariation = getVariation(dump['refinedPeriod'],'Period','us',scale=1e6,printStats=True);
amplitudeVariation = getVariation(dump['angle'],'Amplitude','deg',printStats=True);
blockTempVariation = getVariation(dump['blockTemperature'],'Block temp','C',printStats=True);
boardTempVariation = getVariation(dump['boardTemperature'],'Board temp','C',printStats=True);
pressureVariation = getVariation(dump['pressure'],'Air pressure','hPa',scale=1e-2,printStats=True);
relhumidVariation = getVariation(dump['humidity'],'Relative humidity','%',printStats=True);
```

```python
# Superimpose two hours of period and amplitude data separated by 24 hours, which shows that:
# - The dominant variations (due to gear train errors) are periodic on this timescale.
# - The period varies most over a few minutes while the amplitude variations are slower.
nhour = 3600;
t1,t2 = 2*nhour,26*nhour;
thour = 60.*np.arange(nhour)/nhour;
plt.figure(figsize=(12,10));
plt.subplot(2,1,1);
plt.plot(thour,periodVariation[t1:t1+nhour],'-',label='hour-1');
plt.plot(thour,periodVariation[t2:t2+nhour],'-',label='hour-2');
plt.xlabel('Elapsed time (mins)');
plt.ylabel('Period variation (us)');
plt.legend();
plt.subplot(2,1,2);
plt.plot(thour,amplitudeVariation[t1:t1+nhour],'-',label='hour-1');
plt.plot(thour,amplitudeVariation[t2:t2+nhour],'-',label='hour-2');
plt.xlabel('Elapsed time (mins)');
plt.ylabel('Amplitude variation (deg)');
plt.legend();
```

```python
# Perform a running average of a signal. Use odd length values to avoid half-sample shift.
# Uses decreasing averaging at the edges to preserve the total number of samples.
def runningAvg(signal,length):
    kernel = np.ones((length,))/length
    smooth = np.copy(signal)
    i1 = (length-1)//2
    i2 = len(signal) - length + i1 + 1
    smooth[i1:i2] = np.convolve(signal,kernel,mode='valid')
    return smooth
```

```python
# Smooth environmental parameters with a 55-sec running average.
nsmooth = 55;
blockTempSmooth = runningAvg(dump['blockTemperature'],nsmooth);
boardTempSmooth = runningAvg(dump['boardTemperature'],nsmooth);
pressureSmooth = runningAvg(dump['pressure'],nsmooth);
relhumidSmooth = runningAvg(dump['humidity'],nsmooth);
```

```python
# Plot an hour of environmental variations with and without smoothing, which shows that:
# - Board is about 7 degC warmer than block, presumably due to local heating.
# - Block temperature is more stable than board temperature, as expected due to
#   block's increased thermal mass.
# - Honeywell sensors have some readout aliasing (horizontal banding), but this is
#   effectively removed by the running average.
t1 = 6*nhour;
plt.figure(figsize=(12,12));
plt.subplot(2,2,1);
plt.plot(thour,dump['blockTemperature'][t1:t1+nhour],'r.');
plt.plot(thour,blockTempSmooth[t1:t1+nhour],'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Block Temperature (C)');
plt.subplot(2,2,2);
plt.plot(thour,dump['boardTemperature'][t1:t1+nhour],'r.');
plt.plot(thour,boardTempSmooth[t1:t1+nhour],'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Board Temperature (C)');
plt.subplot(2,2,3);
plt.plot(thour,dump['pressure'][t1:t1+nhour]-1e5,'r.');
plt.plot(thour,pressureSmooth[t1:t1+nhour]-1e5,'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Air Pressure - 100,000 (Pa)');
plt.subplot(2,2,4);
plt.plot(thour,dump['humidity'][t1:t1+nhour],'r.');
plt.plot(thour,relhumidSmooth[t1:t1+nhour],'b-');
plt.xlabel('Elapsed Time (mins)');
plt.ylabel('Relative Humidity (%)');
```

```python
# Use the smoothed block temperature as our best guess of the ambient air temperature.
airTemperature = blockTempSmooth;
```

```python
# Convert the smoothed relative humidity signal into a partial pressure of water vapor.
# The first step is to calculate the saturated water vapor pressure (in Pa) as a function
# of the ambient air pressure and temperature (see http://en.wikipedia.org/wiki/Relative_humidity)
saturatedVaporPressure = 1e2*(1.0007 + 3.46e-8*pressureSmooth)*6.1121*np.exp(17.502*airTemperature/(240.97+airTemperature));
# Next, we scale the saturated pressure by the relative humidity percentage.
vaporPressure = 1e-2*relhumidSmooth*saturatedVaporPressure;
# Plot the results.
plt.figure(figsize=(12,5));
plt.plot(tdays,saturatedVaporPressure,label='saturated');
plt.plot(tdays,vaporPressure,label='actual');
plt.xlabel('Elapsed Time (days)');
plt.ylabel('Water vapor pressure (Pa)');
plt.ylim((0,4000));
plt.legend();
```

```python
# Calculate the dry air partial pressure as the difference between the total and vapor pressures,
# assuming that air can be treated as an ideal gas.
dryPressure = pressureSmooth - vaporPressure;
# Values of the specific gas constants Rx = kB/mx for dry air and water vapor in J/(kg K)
Rdry,Rvapor = 287.058,461.495;
# Calculate the air density using the ideal gas law.
airDensity = (dryPressure/Rdry + vaporPressure/Rvapor)/(airTemperature+273.15);
# Plot the results.
plt.figure(figsize=(12,5));
plt.plot(tdays,airDensity);
plt.xlabel('Elapsed Time (days)');
plt.ylabel('Air Pressure (kg/m^3)');
```

```python
# Model the train errors as an arbitrary additive signal with a period of exactly 12*60*60 samples.
# Since the signal is arbitrary, each of the 12*60*60 possible phases is described by an independent
# parameter, so our train error feature matrix T selects the appropriate parameter for each sample.
trainErrorPeriod = 12*60*60;
ntrainCycles = (nSamples+trainErrorPeriod-1)//trainErrorPeriod;
T = np.tile(np.identity(trainErrorPeriod),(ntrainCycles,))[:,:nSamples];
```

```python
# Use variations in the air density, water vapor pressure, and smoothed block temperature to model
# environmental influcences on the period and amplitude.
airDensityVariation = getVariation(airDensity);
vaporPressureVariation = getVariation(vaporPressure);
blockTempSmoothVariation = getVariation(blockTempSmooth);
X = np.vstack((T,airDensityVariation,vaporPressureVariation,blockTempSmoothVariation)).T;
```

```python
slen=450;
nsmooth=len(periodVariation)//slen;
smooth = lambda y: np.mean(y[:slen*nsmooth].reshape((nsmooth,slen)),axis=1);
periodSmooth = smooth(periodVariation);
tempSmooth = smooth(tempVariation);
pressSmooth = smooth(1e-3*pressVariation);
humidSmooth = smooth(humidVariation);
tRawDays = np.arange(slen*nsmooth)/(24*3600.);
tSmoothDays = smooth(tRawDays);
```

```python
assert 12*3600 % slen == 0, 'slen = %d does not evenly divide 12 hours' % slen
ncycle = 12*3600/slen;
n12hr = (nsmooth+ncycle-1)//ncycle;
T = np.tile(np.identity(ncycle),(n12hr,))[:,:nsmooth];
```

```python
model = linear_model.LinearRegression()
X = np.vstack((tempSmooth,pressSmooth,humidSmooth,T)).T;
model.fit(X,periodSmooth);
periodFit = model.predict(X);
tempCoef,pressCoef,humidCoef = model.coef_[:3];
XT = np.copy(X);
XT[:,:3] = 0;
hourHandVariations = model.predict(XT);
print 'Temperature coef = %.3f ppm/C' % tempCoef
print 'Pressure coef = %.3f ppm/kPa' % pressCoef
print 'Humidity coef = %.3f ppm/%%' % humidCoef
```

```python
plt.figure(figsize=(12,5));
plt.plot(tSmoothDays,periodSmooth,'k+',label='data');
plt.plot(tSmoothDays,periodFit,'k-',label='fit');
plt.plot(tSmoothDays,tempCoef*tempSmooth,'--',label='temp');
plt.plot(tSmoothDays,pressCoef*pressSmooth,'--',label='press');
plt.plot(tSmoothDays,humidCoef*humidSmooth,'--',label='humid');
#plt.plot(tSmoothDays,hourHandVariations,'-',label='12hr')
plt.legend();
plt.xlabel('Elapsed Time (days)');
plt.ylabel('Smoothed period - 2 (us)');
```
