```python
from pymongo import MongoClient
from pymongo import DESCENDING
from pymongo import ASCENDING
```

```python
# Connect to Mongod (fails and exits if mongod is not running)
client = MongoClient('localhost', 4321)
dbname = "TickTock"
dataCollectionName = "datapacketmodels"
bootCollectionName = "bootpacketmodels"


# Get db object
db = client[dbname]

# Get collection object
data = db[dataCollectionName]
boot = db[bootCollectionName]


# Set query parameters
fetch_limit = 1;
# Set fetch constants
ID = u"_id"
RAW = u"raw"
TIMESTAMP = u"timestamp"
COMPUTER_TIMESTAMP = u"computerTimestamp"
CRUDE_PERIOD = u"crudePeriod"
BOARD_TEMPERATURE = u"boardTemperature"
PERIOD = u"refinedPeriod"
```

```python
GPS_EPOCH = 315964800
SECONDS_PER_WEEK = 7*24*60*60

def unix_time(dt):
    epoch = datetime.datetime.utcfromtimestamp(0)
    delta = dt - epoch
    return delta.total_seconds()

def unix_time_millis(dt):
    return unix_time(dt) * 1000.0

# Get initial date
results = boot.find().sort(TIMESTAMP, DESCENDING).limit(1)

for document in results:
    d = datetime.datetime.fromtimestamp(GPS_EPOCH + document['initialWeekNumber']*SECONDS_PER_WEEK + \
                               document['initialTimeOfWeek'] + 7*60*60) #Add UTC-PT offset!
    
print d
print unix_time_millis(d)
```

```python
# Clock Frequency Information
f_cpu = 10000000.
adc_div = 64.
cycles = 13.
timePerTick = adc_div*cycles/f_cpu

readFromFile = False


if readFromFile:
    runningDates = np.loadtxt('runningDates.np')
    computerDates = np.loadtxt('computerDates.np')
    gpsDates = np.loadtxt('gpsDates.np')
else:
    # Perform Query
    results = data.find({TIMESTAMP : {"$gt" : d}},{ID:False,TIMESTAMP:True, COMPUTER_TIMESTAMP:True, \
                PERIOD:True, CRUDE_PERIOD:True}) \
                .sort(TIMESTAMP, ASCENDING) \
#                .limit(2000) \

    totalADCTicks = 0
    deltas = []
    runningDates = np.array([],dtype=uint32)
    computerDates = np.array([],dtype=uint32)
    gpsDates = np.array([],dtype=uint32)
    runningDate = d
    for document in results:
        try:
            deltas.append(document[TIMESTAMP] - (document[COMPUTER_TIMESTAMP] + datetime.timedelta(seconds=16))) # add 16 seconds
            runningDate += datetime.timedelta(seconds=timePerTick*document[CRUDE_PERIOD])
            runningDates = np.append(runningDates, unix_time(runningDate))
            computerDates = np.append(computerDates, unix_time(document[COMPUTER_TIMESTAMP]))
            gpsDates = np.append(gpsDates, unix_time(document[TIMESTAMP]))
            totalADCTicks += document[CRUDE_PERIOD]
        except Exception:
            print document
        
    print str(results.count()) + " results"
```

```python
# Calculate time from ADC ticks
timePassed = totalADCTicks * timePerTick
print timePassed

# print d + datetime.utcfromtimestamp(timePassed)
```

```python
fig = plt.figure(figsize=(12,4),dpi=100)
ax3 = fig.add_subplot(111)

# ax3.plot([r for r in diff(computerDates)])
ax3.plot(diff(gpsDates), color='orange', label="Node Derived Timestamps", alpha=0.8)
ax3.plot(diff(runningDates), color='red', label="Database Derived Timestamps", alpha=0.4)
ax3.set_title("Numpy's array diff for timestamps")
plt.legend()
plt.show()
```

As shown in this graph, node is apparently rounding off each timestamp to the nearest milisecond.  In the documentation, the date object only internally stores the time to milisecond precision.  Because of the truncation the derived timestamp drifts upwards.  The solution is to store the microseconds as well.

```python
rgslope = np.polyfit(range(0,len(runningDates-gpsDates)),runningDates-gpsDates,1)
print "Running - GPS Slope: " + str(rgslope[0]*1000) + " ms/s\t\tIntercept: " + str(rgslope[1])

cgslope = np.polyfit(range(0,len(computerDates-gpsDates)),computerDates-gpsDates,1)
print "Computer - GPS Slope: " + str(cgslope[0]*1000) + " ms/s\t\tIntercept: " + str(cgslope[1])

crslope = np.polyfit(range(0,len(computerDates-runningDates)),computerDates-runningDates,1)
print "Computer - Running Slope: " + str(crslope[0]*1000) + " ms/s\t\tIntercept: " + str(crslope[1])
```
